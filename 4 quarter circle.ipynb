```python
from mplturtle import*
```

```python
def draw_quarter_circle_1(size):
    for i in range(90):
        forward(size)
        penup()
        left(180)
        forward(size)
        left(180)
        pendown()
        left(1)
        
def draw_quarter_circle_2(size):
    pencolor('red')
    for i in range(90):
        forward(size)
        penup()
        left(180)
        forward(size)
        left(180)
        pendown()
        left(1)
        
def draw_quarter_circle_3(size):
    pencolor('blue')
    for i in range(90):
        forward(size)
        penup()
        left(180)
        forward(size)
        left(180)
        pendown()
        left(1)
        
def draw_quarter_circle_4(size):
    pencolor('green')
    for i in range(90):
        forward(size)
        penup()
        left(180)
        forward(size)
        left(180)
        pendown()
        left(1)
        
def draw_circle(size):
    draw_quarter_circle_1(size)
    draw_quarter_circle_2(size)
    draw_quarter_circle_3(size)
    draw_quarter_circle_4(size)
        
```

```python
reset()
draw_circle(60)
```
