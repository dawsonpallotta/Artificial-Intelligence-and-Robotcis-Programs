```python
from mplturtle import *
def move_over(size):
    penup()
    forward(size)
    pendown()
def draw_house(size):
    for i in range(3):
        forward(size)
        left(120)
    for i in range(4):
        forward(size)
        right(90)
def draw_3houses(size):
    for i in range(3):
        draw_house(size)
        move_over(50)
```

```python
reset()
draw_3houses(30)
```
