```python
from mplturtle import *
def move_up(size):
    penup()
    left(90)
    forward(size)
    right(90)
    pendown()
def move_down(size):
    penup()
    right(90)
    forward(size)
    left(90)
    pendown()
def move_over(size):
    penup()
    forward(size)
    pendown()
def draw_house(size):
    for i in range(3):
        forward(size)
        left(120)
    for i in range(4):
        forward(size)
        right(90)
def draw_4houses(size):
    for i in range(4):
        draw_house(size)
        move_over(50)
    move_up(60)
    move_over(-50)
    draw_house(size)
    move_up(60)
def draw_houses2(size):
    for i in range(4):
        draw_house(size)
        move_over(-50)
    move_down(60)
    move_over(50)
    draw_house(size)
    move_down(60)
```

```python
reset()
move_over(-75)

draw_4houses(30)
draw_house(30)
draw_houses2(30)
```
