```python
from mplturtle import*
```

```python
def third_circle_1(size):
    left(90)
    for i in range(120):
        forward(size)
        penup()
        left(180)
        forward(size)
        left(180)
        pendown()
        right(1)
        
def third_circle_2(size):
    pencolor('red')
    for i in range(120):
        forward(size)
        penup()
        left(180)
        forward(size)
        left(180)
        pendown()
        right(1)
        
def third_circle_3(size):
    pencolor('blue')
    for i in range(120):
        forward(size)
        penup()
        left(180)
        forward(size)
        left(180)
        pendown()
        right(1)
        
        
def draw_circle(size):
    third_circle_1(size)
    third_circle_2(size)
    third_circle_3(size)
```

```python
reset()
draw_circle(60)
```
