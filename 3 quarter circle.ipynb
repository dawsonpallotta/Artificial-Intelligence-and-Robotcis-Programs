```python
from mplturtle import*
```

```python
def quarter3_circle(size):
    for i in range(270):
        forward(size)
        penup()
        right(180)
        forward(size)
        right(180)
        pendown()
        right(1)
```

```python
reset()
quarter3_circle(60)
```
